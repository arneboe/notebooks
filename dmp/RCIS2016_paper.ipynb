#### Move goal bending

```python
def move_goal(use_scaling):
    plt.plot(times, positions, "r", label="Original Trajectory")

    lastPhaseValue = 0.1
    numWeights = 100
    overlap = 0.2
    endTime = executionTime
    endPos = positions[-1]


    cs = CS(executionTime, lastPhaseValue)
    dmp = DmpWithImitation(executionTime, positions[0],
                           (positions[1] - positions[0]) / dt, positions[-1],
                           cs, numWeights, overlap, use_scaling)
    dmp.imitate(times, positions, dt)

    cs2 = CS(executionTime, lastPhaseValue)
    dmp.reset(cs2, positions[-1] + 10, executionTime, positions[0])
    (ts, ys, yds) = dmp.run(0.001)
    plt.plot(ts, ys, "m--", label="Trajectories with shifted goal positions")

    for i in [20, 30,40, 50, -10, -20, -30, -40, -50]:
        cs2 = CS(executionTime, lastPhaseValue)
        dmp.reset(cs2, positions[-1] + i, executionTime, positions[0])
        (ts, ys, yds) = dmp.run(0.001)
        plt.plot(ts, ys, "m--")





    #ax2.plot(ts, yds, "c--")
    plt.xlabel('Time in Seconds', fontsize=22)
    plt.ylabel('X-Position of Kick Trajectory', fontsize=22)
    plt.legend(loc=3)
```
